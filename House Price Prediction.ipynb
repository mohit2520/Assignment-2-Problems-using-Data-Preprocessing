```python
df_whole['Utilities'] = df_whole['Utilities'].fillna("AllPub")

df_whole['Electrical'] = df_whole['Electrical'].fillna("SBrkr")

df_whole['Exterior1st'] = df_whole['Exterior1st'].fillna("VinylSd")

df_whole['Exterior2nd'] = df_whole['Exterior2nd'].fillna("VinylSd")
```

```python
# Missing interger values replace with the median in order to return an integer
df_whole['BsmtFullBath']= df_whole.BsmtFullBath.fillna(df_whole.BsmtFullBath.median())
df_whole['BsmtHalfBath']= df_whole.BsmtHalfBath.fillna(df_whole.BsmtHalfBath.median())
df_whole['GarageCars']= df_whole.GarageCars.fillna(df_whole.GarageCars.median())

# Missing float values were replaced with the mean for accuracy 
df_whole['BsmtUnfSF']= df_whole.BsmtUnfSF.fillna(df_whole.BsmtUnfSF.mean())
df_whole['BsmtFinSF2']= df_whole.BsmtFinSF2.fillna(df_whole.BsmtFinSF2.mean())
df_whole['BsmtFinSF1']= df_whole.BsmtFinSF1.fillna(df_whole.BsmtFinSF1.mean())
df_whole['GarageArea']= df_whole.GarageArea.fillna(df_whole.GarageArea.mean())
df_whole['MasVnrArea']= df_whole.MasVnrArea.fillna(df_whole.MasVnrArea.mean())
```

```python
df_whole.GarageYrBlt.fillna(df_whole.YearBuilt, inplace=True)
df_whole.TotalBsmtSF.fillna(df_whole['1stFlrSF'], inplace=True)

sns.lmplot(x="TotalBsmtSF", y="1stFlrSF", data=df_whole)
plt.title("Linear Regression of Basement SF and 1rst Floor SQ ")
plt.xlim(0,)
plt.ylim(0,)
plt.show()
```

```python
df_whole['LotFrontage']= df_whole.LotFrontage.fillna(np.sqrt(df_whole.LotArea))
df_whole['LotFrontage']= df_whole['LotFrontage'].astype(int)
```

```python
countmissing = df_whole.isnull().sum().sort_values(ascending=False)
percentmissing = (df_whole.isnull().sum()/df_whole.isnull().count()).sort_values(ascending=False)
wholena = pd.concat([countmissing,percentmissing], axis=1)
wholena.head(3)
```

```python
Livingtotalsq = df_whole['TotalBsmtSF'] + df_whole['1stFlrSF'] + df_whole['2ndFlrSF'] + df_whole['GarageArea'] + df_whole['WoodDeckSF'] + df_whole['OpenPorchSF']
df_whole['LivingTotalSF'] = Livingtotalsq

# Total Living Area divided by LotArea
df_whole['PercentSQtoLot'] = df_whole['LivingTotalSF'] / df_whole['LotArea']

# Total count of all bathrooms including full and half through the entire building
df_whole['TotalBaths'] = df_whole['BsmtFullBath'] + df_whole['BsmtHalfBath'] + df_whole['HalfBath'] + df_whole['FullBath']

# Percentage of total rooms are bedrooms
df_whole['PercentBedrmtoRooms'] = df_whole['BedroomAbvGr'] / df_whole['TotRmsAbvGrd']

# Number of years since last remodel, if there never was one it would be since it was built
df_whole['YearSinceRemodel'] = 2016 - ((df_whole['YearRemodAdd'] - df_whole['YearBuilt']) + df_whole['YearBuilt'])
```

```python
pricing1 = df_whole[['Id','SalePrice','MiscVal']]

neigh = df_whole[['Neighborhood','MSZoning','MSSubClass','BldgType','HouseStyle']]

dates = df_whole[['YearBuilt','YearRemodAdd','GarageYrBlt','YearSinceRemodel']]

quacon = df_whole[['ExterQual','BsmtQual','PoolQC','Condition1','Condition2','SaleCondition',
                  'BsmtCond','ExterCond','GarageCond','KitchenQual','GarageQual','HeatingQC','OverallQual','OverallCond']]

features =  df_whole[['Foundation','RoofStyle','RoofMatl','Exterior1st','Exterior2nd',
                     'MiscFeature','PavedDrive','Utilities',
                     'Heating','CentralAir','Electrical','Fence']]

sqfoot = df_whole[['LivingTotalSF','TotalBsmtSF','1stFlrSF','2ndFlrSF','GrLivArea',
                  'GarageArea','WoodDeckSF','OpenPorchSF','LotArea','PercentSQtoLot','LowQualFinSF']]

roomfeatcount = df_whole[['PercentBedrmtoRooms','TotalBaths','FullBath','HalfBath',
                         'KitchenAbvGr','TotRmsAbvGrd','Fireplaces','GarageCars','GarageType','EnclosedPorch']]

# Splits out sale price for the training set and only has not null values
pricing = df_whole['SalePrice']
pricing = pricing[pricing.notnull()]

# Bringing it all together
df_whole = pd.concat([pricing1,neigh,dates,quacon,features,sqfoot,roomfeatcount], axis=1)
```

```python
df_whole = pd.get_dummies(df_whole)
```

```python
train_X = df_whole[df_whole['SalePrice'].notnull()]
del train_X['SalePrice']
test_X =  df_whole[df_whole['SalePrice'].isnull()]
del test_X['SalePrice']
```

```python
# Split the data into 75% training  and 25% testing
from sklearn.model_selection import train_test_split
train_valid_X = train_X
train_valid_y = pricing
test_X = test_X
train_X , valid_X , train_y , valid_y = train_test_split( train_valid_X , train_valid_y , train_size = .7 ,random_state = 0 )
```

```python
from sklearn.linear_model import Ridge
model = Ridge()

# Use random forest classifier 
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators = 10, criterion='entropy',random_state =0)
forest.fit(train_X , train_y )
```

```python
# Get the accuracy on the training data set
forest.score(train_X , train_y)
```
